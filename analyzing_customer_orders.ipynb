Analyzing Customer Orders Using Python

Task : Use Python DS to classify products, identify customer purchase patters and generate business insights

```python
# 1. Store customer orders
# create a list of customer names
customer_names = ["Alice", "Charlie", "John", "Bob", "Mary", "Philip"]

# store customer order details ( customer name, product, price, category ) 
order_details = [("Alice", "Laptop", 600, "Electronics"),
                ("Charlie", "Mobile", 550, "Electronics"),
                ("John", "Cooking Range", 800, "Home Essentials"),
                ("Bob", "T-shirt", 10, "Clothing"),
                ("Mary", "Vaccum cleaner", 100, "Home Essentials"),
                ("Alice", "Blazer", 100, "Clothing"),
                ("John", "Power Bank", 100, "Electronics"),
                ("Philip", "Mobile", 400, "Electronics"),
                ("Philip", "Pants", 30, "Clothing")]

# use a dictionary where keys are customer names and values are list of ordered products
customer_orders = {}
for name, product, _, _ in order_details :
    if name not in customer_orders :
        customer_orders[name] = []
    customer_orders[name].append(product)
print("\nCustomer orders : \n" , customer_orders)
```

```python
# 2. Classify products by category
# map each product to its category
product_category_map = {}
for _, product, _, category in order_details:
    product_category_map[product]=category
print("\nProduct : Category = \n" , product_category_map)

# create a set of unique product categories
product_categories = set(product_category_map.values())

# Display all available product categories
print("\nUnique Product categories: \n", product_categories)
```

```python
# 3. Analyze customer orders

# calculate total amount each customer spends
customer_total = {}
for name,_,price,_ in order_details:
    if name not in customer_total:
        customer_total[name] = 0
    customer_total[name] += price
print("\nTotal spend by each customer : \n", customer_total)

# classify customers as high-value buyer, moderate-buyer, low-value buyer
customer_classification={}
for name, total in customer_total.items():
    if total > 100:
       customer_classification[name] = "High Value Buyer"
    elif total > 50 and total <= 100:
        customer_classification[name] = "Moderate Buyer"
    else:
        customer_classification[name] = "Low Value Buyer"   
print("\nCustomer Classification :\n" ,customer_classification)
```

```python
# 4. Generate business insights

# calculate total revenue per product category 
category_revenue = {}
for _,_, price, category in order_details:
    if category not in category_revenue:
        category_revenue[category] = 0
    category_revenue[category] += price
print("\nTotal revenue per product category : \n" , category_revenue)

# extract unique products from all orders
unique_products = {products for _,products,_,_ in order_details}
print("\nUnique products from all order : \n",unique_products)

# Use a list comprehension to find all customers who purchased electronics
cust_pur_electronics = [name for name, _, _, category in order_details if category == "Electronics"]
print("\nCustomers who purchased elecronics : \n",cust_pur_electronics)

# Identify the top three highest spending customers using sorting
sorted_total_spend_list = sorted(customer_total.items(), key=lambda x:x[1], reverse=True)
top_three_customers = sorted_total_spend_list[:3]
print("\nTop three highest spending customers: \n", top_three_customers)
```

```python
# 5. Organize and display data

# print a summary of each customer's total spending and their classification
print("\nSummary of customer's total spending and their classification :")
for name in customer_total.keys():
    print(f"{name} spend ${customer_total[name]} - {customer_classification[name]}")

#Find customers who purchased from multiple categories
customer_purchase_cate = {}
for name,_,_,category in order_details:
    if name not in customer_purchase_cate:
        customer_purchase_cate[name] = []
    customer_purchase_cate[name].append(category)
multi_category_customers = { 
    person 
    for person, category in customer_purchase_cate.items()
    if len(set(category)) > 1   
}
print("\nCustomers who purchased from multiple categories: \n", multi_category_customers)

# Identify common customers who bought both electronics and clothing
electronics = set()
clothing = set()
for name,categories in customer_purchase_cate.items():
    if "Electronics" in categories:
        electronics.add(name)
    if "Clothing" in categories:
        clothing.add(name)
print(f"\nCommon customers who brought both electronics and clothing are : \n", electronics & clothing)
```
